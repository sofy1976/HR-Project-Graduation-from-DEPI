```python
#Create a new Column called AgeGroup to Divide Employees into 2 Categories(Below_Average and Above_Avergae)
def Age_group(x):
    if x<29:
        return 'Below_Average'
    else:
        return 'Above_Avergae'
```

```python
Emp_merge['AgeGroup']=Emp_merge['Age'].apply(Age_group)
```

```python
# Dropping Unneecessary Columns in Emp_merge DataFrame
Emp_merge.drop(columns=['FirstName','LastName','Education','HireDate'],inplace=True)
```

```python
# Fix the problem with EducationField where "Marketing" is repeated because of a space at the end of some entries
def fix_Marketing(x):
    if x=='Marketing ':
        return 'Marketing'
    else:
        return x
```
